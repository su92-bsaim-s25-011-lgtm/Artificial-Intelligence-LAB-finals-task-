```python
#dropping rows where target column is missing
data= data.dropna(subset= ['RainTomorrow'])
```

```python
data= data.drop_duplicates()
```

```python
#Dropping unnecessary columns/ columns with high number of missing values
columns_to_drop = [
    'WindDir9am',
    'WindDir3pm',
    'WindGustDir',
    'Evaporation',
    'Sunshine',
    'Cloud9am',
    'Cloud3pm',
]
data.drop(columns=columns_to_drop, axis=1, inplace=True, errors='ignore')
```

```python
#feature extraction from Date column
data['Date']= pd.to_datetime(data['Date'])
data['month']= data['Date'].dt.month
data['day']= data['Date'].dt.day
data= data.drop(columns=['Date'])  #after feature extraction, dropped the column
```

```python
#fill missing values
#numeric columns with median (robusting to outliers)
numcols= data.select_dtypes(include=['float64', 'int64']).columns
data[numcols]= data[numcols].fillna(data[numcols].median())

#categorical columns with mode
catcols= data.select_dtypes(include=['object']).columns
for col in catcols:
  data[col]= data[col].fillna(data[col].mode()[0])
print("missing values:", data.isnull().sum().sum())
```

```python
#label encoding (converting categorical data into numerical data)
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()

obj= data.select_dtypes(include='object').columns
for col in obj:
  data[col]= le.fit_transform(data[col])
```

```python
#confirm no nulls remain and all columns are numeric
print('Shape: ', data.shape)
print("Null values:", data.isnull().sum().sum())
data.dtypes
```

```python
#check unique values of location to confirm that encoding worked
print('Unique location values: ', data['Location'].unique())
```

```python
#creating variables x and y for prediction
x= data.drop('RainTomorrow', axis=1)
y= data['RainTomorrow']

print('x shape:', x.shape)
print('y shape:', y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(
    x,y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.ensemble import RandomForestClassifier

RF= RandomForestClassifier(
    n_estimators=100,
    max_depth=15,
    min_samples_leaf=5,
    class_weight='balanced',
    random_state=42
)

RF.fit(x_train,y_train)
```

```python
#make predictions
y_pred= RF.predict(x_test)
y_prob= RF.predict_proba(x_test)[:, 1] #probability of rain
```

```python
import sklearn.metrics as metrics
accuracy= metrics.accuracy_score(y_test, y_pred)
precision= metrics.precision_score(y_test, y_pred, average='binary')
recall= metrics.recall_score(y_test, y_pred, average='binary')
f1= metrics.f1_score(y_test, y_pred, average='binary')
roc_auc= metrics.roc_auc_score(y_test, y_prob)

print('Model performance metrics:')
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'ROC AUC Score: {roc_auc:.4f}')
```

```python
#Save the model
import pickle

with open('model.pkl', 'wb') as f:
    pickle.dump(RF, f)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

data_reg = data.copy()
data_reg['MaxTemp_Tomorrow']= data_reg['MaxTemp'].shift(-1)
data_reg= data_reg.dropna(subset=['MaxTemp_Tomorrow'])

x_reg=data_reg.drop(columns=['RainTomorrow', 'MaxTemp_Tomorrow'])
y_reg= data_reg['MaxTemp_Tomorrow']
```

```python
#train/test split for regression
X_train, X_test, Y_train, Y_test = train_test_split(
    x_reg, y_reg,
    test_size= 0.2,
    random_state=42
)

print('Regression training samples:', X_train.shape[0])
print('Regression testing samples:', X_test.shape[0])
```

```python
lr= LinearRegression()
lr.fit(X_train, Y_train)
```

```python
y_pred_r= lr.predict(X_test)
mae= mean_absolute_error(Y_test, y_pred_r)
rmse= mean_squared_error(Y_test, y_pred_r) ** 0.5
r2= r2_score(Y_test, y_pred_r)

print('MaxTemp prediction performance:')
print(f'MAE (mean absolute error): {mae:.2f}')
print(f'RMSE (root mean square error): {rmse:.2f}')
print(f'R2 Score (accuracy of fit): {r2:.4f}')
```

```python
with open('lr_model.pkl', 'wb') as f:
    pickle.dump(lr, f)
```
